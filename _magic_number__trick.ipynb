```python
import math

# e.g. k = 3 and e = 56, then we wish to find 0.056
# then we output a list of integer m
def ml2(s2, n, k, e):
  s2 = s2 - math.floor(s2)
  lower_bound = e / 10**k
  upper_bound = lower_bound + 10**(-k)
  u, list1 = 0, []
  for i in range(n):
    if lower_bound <= u < upper_bound: list1.append(i)
    u += s2
    if u >= 1: u -= 1
  return list1

test = ml2(math.sqrt(2), 10**6, 5, 1001)
hide_value = 77 * math.sqrt(2)
print(hide_value)
print(hide_value * test[0])
# in the above example, hide_value = 77 x sqrt(2) = 108.89444430272833
# but if the 108.89444430272833 x 7979 = 868868.7710914693
# where it can appear the same integer 77 after the decmial point .77
```
